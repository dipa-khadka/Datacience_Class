```python
alpha = 0.05 # unless stated it's value is 0.05
```

```python
# say i toss ten times coin toss so
n = 10
```

```python
no_of_heads = range(0, n+1)
```

```python
from scipy.special import comb
prob =[]
p=0.5
q=1-p
for x in no_of_heads:
    px = comb(N=n, k=x)*p**x * q**(n-x)
    prob.append(px)
```

```python
# bar plotting
import matplotlib.pyplot as plt
```

```python
# difinig size of figure
fig,ax=plt.subplots(figsize=(20,8))
ax.bar(x=no_of_heads, height=prob)
ax.bar_label(ax.containers[0])
ax.set_xticks(no_of_heads)
fig.show()
```
