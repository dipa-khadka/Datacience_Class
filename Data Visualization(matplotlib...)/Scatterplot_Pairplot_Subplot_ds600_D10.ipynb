# Scatterplot:
A scatter plot (aka scatter chart, scatter graph) uses dots to represent values for two different numeric variables.
   It identifies a possible relationship between changes observed intwo different setsof variables.
Scatter plots are used to observe relationships between variables.


```python
 import seaborn as sns
```

```python
# scatterplot shows relationship between two variables.
```

```python
df = sns.load_dataset('iris')
```

```python
df.head(2)
```

```python
import matplotlib.pyplot as plt
```

```python
# -------------Petal_length & petal_weidth--------------
plt.scatter(x=df['petal_length'], y=df['petal_width'])
plt.xlabel('petal_length')
plt.ylabel('ppetal_width')
plt.show()
```

The figure shows as the value of petal length increases the value of petal width also increases. It shows linear relationship between them. They both rise and fall together . This is also called correlation.

```python
sns.scatterplot(x=df['petal_length'], y=df['petal_width'])
#  in scatter we need to name x and y label but in scatterplot it automatically name by it self
```

```python
# we can see correlation between continuous data using df.corr() method.
df.corr(numeric_only=True)
```



> correlation value ranges from [-1,1]:
* if correlation is near to -1; It is negative correlation which means one rises other falls
* if correlation is near to 0; It is zero correlation which means there is no relationship
* if correlation is near to 1; It is positive correlation which means they both rise fall together.



```python
# you can plot this correlation value in heatmap
sns.heatmap(df.corr(numeric_only=True))
```

```python
# yesma chahi hamilai garo hunxa kun ko kati value xa and to find out which relation they share between them
#  so making easy too find out the relation between them we  just add annot= True
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
```

# Multivariate Scatterplot


```python
df.sample(3, random_state=40)
```

```python
# we can use hue parameter to color  code the scatterplot based on category/species
```

```python
sns.scatterplot(x=df['petal_length'], y=df['petal_width'], hue=df['species'])
```

# Pairplot:


```python
sns.pairplot(df, hue='species')
```

```python
#  1. setosa species ko matra data tanney ani tyo taneko data ko petal_width access garne
#  2. versicolor species ko matra data tanney ani tyo taneko data ko petal_width access garne
#  3. virginica species ko matra data tanney ani tyo taneko data ko petal_width access garne
#  4. sabai ko kdeplot garne
```

```python
setosa_pw = df.loc[df.species == 'setosa']['petal_width']
versicolor_pw = df.loc[df.species == 'versicolor']['petal_width']
virginica_pw= df.loc[df.species == 'virginica']['petal_width']

sns.kdeplot(setosa_pw , fill= True)
sns.kdeplot(versicolor_pw, fill= True)
sns.kdeplot(virginica_pw, fill=True)
```

# Subplots

```python
import numpy as np
```

```python
x = np.linspace(-5,5, 100)
```

```python
sinx = np.sin(x)
cosx = np.cos(x)
tanx = np.tan(x)


```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
```

```python
fig, axes=plt.subplots(nrows=2, ncols=2, figsize=(10,5))

ax = axes[0,0]
ax.plot(x, sinx, label='sinx')
ax.legend()

ax = axes[0,1]
ax.plot(x, cosx, label='cosx')
ax.legend()

ax = axes[1,0]
ax.plot(x, tanx, label='tan')
ax.legend()
```
