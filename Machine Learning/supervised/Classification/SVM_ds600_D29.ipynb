```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
df.head(2)
```

```python
# species ko aadharma histogram plot
sns.kdeplot(x='sepal_length', hue='species', data=df, fill=True) # for single variable
plt.show()
```

The sepal_length distribution of different species are overlapping so we can't use sepal length and sepal width to classify flower classes

```python
sns.scatterplot(x='petal_length',y='petal_width', hue='species', data=df)
plt.show()
```

Using petal length and petal width we can classify iris flower classes

```python
# we can visualize all this on pairplot
sns.pairplot(df, hue='species')
```

```python
X = df[['petal_length','petal_width']]
y = df.species
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
```

```python

from sklearn.svm import SVC
```

```python
model = SVC()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation :-

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
ConfusionMatrixDisplay.from_estimator(model, X_test, y_test)
```

```python
from sklearn.metrics import classification_report, f1_score,accuracy_score
print(classification_report(y_test,y_pred))
```

```python
f1_score(y_test, y_pred, average='weighted') # it is a multiclass classification binary ma average = 'binary' default hunxa
```

```python
accuracy_score(y_test, y_pred)
```

## Decision Boundry :-

```python
from sklearn.inspection import DecisionBoundaryDisplay
```

```python
DecisionBoundaryDisplay.from_estimator(model, X_test)
sns.scatterplot(x='petal_length', y='petal_width', hue=y, data=X)
plt.show()
```

```python

# Logistic Regression
from sklearn.linear_model import LogisticRegression
lmodel = LogisticRegression()
lmodel.fit(X_train, y_train)
y_pred = lmodel.predict (X_test)
```

```python
# Evaluation
ConfusionMatrixDisplay.from_estimator(lmodel, X_test, y_test)
```

```python
f1_score(y_test, y_pred,average='weighted')
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
